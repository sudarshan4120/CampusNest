```python
neighbourhood_zipcode_mapping['Zip_Code'] = neighbourhood_zipcode_mapping['Zip_Code'].apply(
    lambda x: list(map(int, x.split(',')))
)
# Display the updated DataFrame
print(neighbourhood_zipcode_mapping)
```

```python
neighbourhood_mapping = neighbourhood_zipcode_mapping.explode('Zip_Code')
neighbourhood_mapping['Zip_Code'] = neighbourhood_mapping['Zip_Code'].astype(int)
```

```python
zipcodes_in_boston = neighbourhood_mapping['Zip_Code'].tolist()
zipcodes_in_rental_listing = rental_listing['Zipcode'].tolist()
zipcodes_in_rental_listing = pd.Series(zipcodes_in_rental_listing).unique().tolist()
zipcodes_not_covered = [item for item in zipcodes_in_boston if item not in zipcodes_in_rental_listing]
neighbourhoods_not_covered = neighbourhood_mapping[neighbourhood_mapping['Zip_Code'].isin(zipcodes_not_covered)]

# Convert the result to a dictionary for easy mapping
zipcode_to_neighbourhood = neighbourhood_mapping.set_index('Zip_Code')['Neighbourhood'].to_dict()

# Display the mapping
print(neighbourhoods_not_covered)
```
